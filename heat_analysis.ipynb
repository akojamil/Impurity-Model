```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

# Center Line of Teflon Circle

```python
data = pd.read_csv("setup4center_clean.csv")
data.head()
```

```python
# Timescale in seconds on which to view the data. 3601 is full-length
TSCALE = 3601
# temperatures only
temp_data = data.drop("x", axis=1).transpose()[0:TSCALE]
```

```python
temp_data.head()
```

```python
# Median temperatures
median_temps = temp_data.median(axis=1)
#median_temps.shape
# Minimum temperatures
min_temps = temp_data.min(axis=1)
max_temps = temp_data.max(axis=1)
```

```python
#median_temps.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(np.arange(TSCALE), max_temps, label="Maximum")
plt.plot(np.arange(TSCALE), median_temps, label="Median")
plt.plot(np.arange(TSCALE), min_temps, label="Minimum")
plt.title("Temperature over Time")
plt.legend(loc=4)
plt.show()
```

```python
max_temps.to_csv("setup4max.csv")
median_temps.to_csv("setup4med.csv")
min_temps.to_csv("setup4min.csv")
```
